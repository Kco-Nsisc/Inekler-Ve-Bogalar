```python
import random
deneme = 0
toplaminek = 0
toplamboğa = 0
sayi = random.randint(1000,9999)
sayiListesi = []
x = 0


for i in str(sayi):
    sayiListesi.append(i)
    sayiListesi[x] = int(sayiListesi[x])
    x = x + 1
    inekSeti = set()
while True:
    girilenSayiListesi = []
    inek = 0
    boğa = 0
    girilenSayi = int(input("Sayı giriniz. "))
    deneme = deneme + 1
    y = 0
    for i in str(girilenSayi):
        girilenSayiListesi.append(i)
        girilenSayiListesi[y] = int(girilenSayiListesi[y])
        y = y + 1
    if girilenSayiListesi[0] == sayiListesi[0]:
        inek = inek + 1
        inekSeti.add(girilenSayiListesi[0])
    elif sayiListesi.count(girilenSayiListesi[0]) == 1:
        inekListesi = list(inekSeti)
        if inekListesi.count(girilenSayiListesi[0]) == 0:
            boğa = boğa + 1
   
    if girilenSayiListesi[1] == sayiListesi[1]:
        inek = inek + 1
        inekSeti.add(girilenSayiListesi[1])
    elif sayiListesi.count(girilenSayiListesi[1]) == 1:
        inekListesi = list(inekSeti)
        if inekListesi.count(girilenSayiListesi[1]) == 0:
            boğa = boğa + 1
    
    if girilenSayiListesi[2] == sayiListesi[2]:
        inek = inek + 1
        inekSeti.add(girilenSayiListesi[2])
    elif sayiListesi.count(girilenSayiListesi[2]) == 1:
        inekListesi = list(inekSeti)
        if inekListesi.count(girilenSayiListesi[2]) == 0:
            boğa = boğa + 1
    
    if girilenSayiListesi[3] == sayiListesi[3]:
        inek = inek + 1
        inekSeti.add(girilenSayiListesi[3])
    elif sayiListesi.count(girilenSayiListesi[3]) == 1:
        inekListesi = list(inekSeti)
        if inekListesi.count(girilenSayiListesi[3]) == 0:
            boğa = boğa + 1
            
    toplaminek += inek
    toplamboğa += boğa
    if sayi == girilenSayi:
        print(f"Tebrikler oyunu kazandınız. Oyunu {deneme} denemede geçtiniz.")
        break
        
    print(f"İnek: {inek} Boğa: {boğa}")
```
